```python
import pandas as pd
#!pip install openpyxl

# File path
file_path = "/Users/treypallace/Desktop/SNPRealTimeVGT.xlsx"


vgt = pd.read_excel(file_path, sheet_name="VGT_monthly")
snp = pd.read_excel(file_path, sheet_name="SNP_monthly")
cpi = pd.read_excel(file_path, sheet_name="CPI")
```

```python
snp.info()
```

```python
import matplotlib.pyplot as plt

# Plot 'Open', 'High', 'Low', and 'Close' against 'Date'
plt.figure(figsize=(10, 6))

plt.plot(snp['Date'], snp['Open'], label='Open', marker='o')
plt.plot(snp['Date'], snp['High'], label='High', marker='o')
plt.plot(snp['Date'], snp['Low'], label='Low', marker='o')
plt.plot(snp['Date'], snp['Close'], label='Close', marker='o')

plt.title('SNP Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
